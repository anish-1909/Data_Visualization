```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("aws_supply_chain_orders_raw.csv")

sns.set_style("whitegrid")
sns.countplot(data=df,x="region")
plt.title("Orders by Region")
plt.show
```

```python
sns.countplot(
    data=df,
    x="warehouse",
    hue="region"
)
plt.show()
```

```python
sns.countplot(data=df,
              x="region",
              palette="bright")

plt.xlabel("region")
plt.ylabel("counts")
plt.legend()
plt.show()
```

```python
sns.barplot(data=df, x="warehouse",y="order_qty",palette="Set3",errorbar=None)
plt.title("averge of Order quantity by warehouse")
plt.xlabel("warehouse")
plt.ylabel("order_qty")

plt.show()
```

```python
sns.lineplot(
    data=df,
    x="order_date",
    y="order_qty"
)

plt.show()
```

```python
sns.histplot(
    data=df,
    x="delivery_time_days",
    bins=5,
    kde=True
)

plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x="order_qty",y="delivery_time_days",hue="status",style="region",alpha=0.7)
plt.title("Order Quantity vs Delivery time")
plt.show()
```

```python
sns.boxplot(
    data=df,
    x="region",
    y="delivery_time_days"
)

plt.show()
```

```python
sns.violinplot(
    data=df,
    x="region",
    y="delivery_time_days")
    
plt.show()
```
