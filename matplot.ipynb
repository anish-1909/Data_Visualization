```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("aws_supply_chain_orders_raw.csv")

orders = df.groupby("region")["order_qty"].sum()
plt.plot(
    orders.index,
    orders.values,
    marker="o",
    linestyle="--"
)

plt.title("Total Orders by Region")
plt.xlabel("Region")
plt.ylabel("Order Quantity")
plt.show()
```

```python
warehouse_orders=df.groupby("warehouse")["order_qty"].sum()
plt.bar(warehouse_orders.index,warehouse_orders.values,color="red",width=0.5)
plt.title("Orders by Warehouse")
plt.xlabel("warehouse")
plt.ylabel("Total Orders")
plt.xticks(rotation=45)
plt.show()
warehouse_orders
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["delivery_time_days"], bins=20)
plt.title("Delivery time distribution")
plt.xlabel("Delivery Days")
plt.ylabel("Frequency")
plt.show()
print(df["delivery_time_days"])
```

```python
plt.scatter(
    df["order_qty"],
    df["delivery_time_days"],
    alpha=0.5
)

plt.title("Order Quantity vs Delivery Time")
plt.xlabel("Order Quantity")
plt.ylabel("Delivery Time")
plt.grid(True)
```

```python
status_counts=df["status"].value_counts()
plt.pie(status_counts.values,labels=status_counts.index,autopct="%1.1f%%")
plt.title("order status distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3]

north = [10,20,15]
south = [5,15,25]

plt.figure(figsize=(8,5))

plt.plot(x, north, label="North")
plt.plot(x, south, label="South")

plt.title("Regional Sales Trend")

plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.legend()

plt.show()
```
